```python
browser_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/browser.tsv',delimiter='\t', header=None, index_col=0, encoding='latin-1')
search_eng_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/search_engines.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
country_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/country.tsv',delimiter='\t', header=None, index_col=0, encoding='latin-1')
event_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/event.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
language_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/languages.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
os_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/operating_systems.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
refer_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/referrer_type.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
connect_dict = pd.read_csv('bmwprod_2017-04-30-lookup_data/connection_type.tsv', delimiter='\t', header=None, index_col=0, encoding='latin-1')
```
